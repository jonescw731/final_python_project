```python
def scrape():
  headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
  for url in urls:
    response = requests.get(url=url, headers=headers, )
    soup = BeautifulSoup(response.text, 'html.parser')
    info = f"Scraping ... {soup.title.string if soup.title else 'No title found'} | {soup.figcaption.string if soup.figcaption else 'Not found'} | {soup.time.string if soup.time else 'Not found'}"
    print(info)
```

```python
scrape()
```

```python
def scrape():
  headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
  for url in urls:
    response = requests.get(url=url, headers=headers, )
    soup = BeautifulSoup(response.text, 'html.parser')
    paragraphs = soup.find_all('p')
    for p in paragraphs:
      article = f"{'\n'.join(p.get_text())}"
      print(article)
```

```python
scrape()
```
